```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/kirtan/Downloads/Sleep_health_and_lifestyle_dataset.csv')
```

```python
df.head()
```

```python
clean_df = df.drop(columns='Person ID', axis=1)
```

```python
clean_df.info()
```

```python
clean_df.shape
```

```python
Categorical_Col = []
Numerical_Col = []
```

```python
for col in clean_df.columns:
    if clean_df[col].dtype == 'object':
        Categorical_Col.append(col)
       
    else:
        Numerical_Col.append(col)
```

```python
Categorical_Col
```

```python
Numerical_Col
```

# Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE = LabelEncoder()
```

```python
for col in Categorical_Col:
    clean_df[col] = LE.fit_transform(clean_df[col])
```

```python
clean_df.head()
```

# Data Visualization

```python
sns.pairplot(clean_df, hue='Sleep Disorder')
plt.show()
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(clean_df.corr(), cmap='jet', linewidths=1, linecolor='black', annot=True)
plt.show()
```

```python
sns.countplot(data=clean_df, x='Sleep Disorder')
plt.show()
```

# Train & Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = clean_df.iloc[:,:-1]
X
```

```python
y = clean_df['Sleep Disorder']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RNF = RandomForestClassifier()
```

```python
RNF.fit(X_train, y_train)
```

Prediction

```python
y_hat = RNF.predict(X_test)
```

# Model Evaluation

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print(accuracy_score(y_test, y_hat))
```

```python
print(classification_report(y_test, y_hat))
```

```python
sns.heatmap(confusion_matrix(y_test, y_hat), cmap='coolwarm', linecolor='white', linewidths=1, annot=True)
plt.show()
```
